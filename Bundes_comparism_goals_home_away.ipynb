```python
data.drop(['Unnamed: 0', 'SEASON', 'HOME_ICON', 'AWAY_TEAM_ID', 'FINISHED', 'LEAGUE', 'VIEWER', 'HOME_TEAM_ID', 'DRAW', 'WIN_HOME', 'WIN_AWAY', 'AWAY_ICON'], axis=1)
```

```python
data['MATCH_DATE'] = pd.to_datetime(data['MATCH_DATE'])
data['MATCH_DATE'] = data['MATCH_DATE'].dt.year
```

```python
data.head()
```

```python
season_goals_sum_HOME = data.groupby('SEASON')['GOALS_HOME'].sum().reset_index()
print(season_goals_sum_HOME)
```

```python
season_goals_sum_AWAY = data.groupby('SEASON')['GOALS_AWAY'].sum().reset_index()
print(season_goals_sum_AWAY)
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,6))
ax[0] = sns.barplot(data=season_goals_sum_HOME, x='SEASON', y='GOALS_HOME', ax=ax[0])
ax[0].set(title='Barplot for home goals', xlabel='Season', ylabel='Home goals')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=90, ha='center')


ax[1] = sns.barplot(data=season_goals_sum_AWAY, x='SEASON', y='GOALS_AWAY', ax=ax[1])
ax[1].set(title='Barplot for home away', xlabel='Season', ylabel='Home Away')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=90, ha='center')
plt.show()
```

```python
mean_goals_home = season_goals_sum_HOME['GOALS_HOME'].mean()
median_goals_home = season_goals_sum_HOME['GOALS_HOME'].median()
max_goals_home = season_goals_sum_HOME['GOALS_HOME'].min()
min_goals_home = season_goals_sum_HOME['GOALS_HOME'].max()
std_goals_home = season_goals_sum_HOME['GOALS_HOME'].std()
print("Mean: ", mean_goals_home)
print("Median: ", median_goals_home)
print("Minimum: ", min_goals_home)
print("Maximum: ", max_goals_home)
print("Standard Deviation: ", std_goals_home)
```

```python
mean_goals_away = season_goals_sum_AWAY['GOALS_AWAY'].mean()
median_goals_away = season_goals_sum_AWAY['GOALS_AWAY'].median()
min_goals_away = season_goals_sum_AWAY['GOALS_AWAY'].min()
max_goals_away = season_goals_sum_AWAY['GOALS_AWAY'].max()
std_goals_away = season_goals_sum_AWAY['GOALS_AWAY'].std()
print("Mean: ", mean_goals_away)
print("Median: ", median_goals_away)
print("Minimum: ", min_goals_away)
print("Maximum: ", max_goals_away)
print("Standard Deviation: ", std_goals_away)
```

```python
mean_home = round(mean_goals_home)
mean_away = round(mean_goals_away)
median_home = round(median_goals_home)
median_away = round(median_goals_away)
max_home = round(max_goals_home)
max_away = round(max_goals_away)
min_home = round(min_goals_home)
min_away = round(min_goals_away)
std_home = round(std_goals_home)
std_away = round(std_goals_away)
print("Mean Home:", mean_home)
print("Mean Away:", mean_away)
print("Median Home:", median_home)
print("Median Away:", median_away)
print("Max Home:", max_home)
print("Max Away:", max_away)
print("Min Home:", min_home)
print("Min Away:", min_away)
print("Std Home:", std_home)
print("Std Away:", std_away)
```

```python
import numpy as np
```

```python
labels = ['Mean', 'Median', 'Max', 'Min', 'Std']
x = np.arange(len(labels))
home_values = [mean_home, median_home, max_home, min_home, std_home]
away_values = [mean_away, median_away, max_away, min_away, std_away]
width = 0.35 
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, home_values, width, label='Goals home')
rects2 = ax.bar(x + width/2, away_values, width, label='Goals away')
ax.set_xlabel('Statistics')
ax.set_ylabel('Goals')
ax.set_title('Comparison of Goals home and Goals away')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```
